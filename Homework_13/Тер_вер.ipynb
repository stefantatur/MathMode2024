```python
# Задание 1 
from scipy.stats import norm

mu = 0  # математическое ожидание
sigma = 4  # стандартное отклонение
p = norm.cdf(6, loc=mu, scale=sigma) - norm.cdf(-6, loc=mu, scale=sigma)

print("Вероятность того, что ошибка взвешивания не превышает 6 граммов:", p)
```

```python
# Задание 2
def probability_wait():
    # Границы интервала времени ожидания автобуса
    a = 0
    b = 5
    probability = (2 - a) / (b - a)
    return probability

probability = probability_wait()
print("Вероятность того, что пассажир будет ждать автобус не более 2 минут:", probability)
```

```python
# Задание 3
def probability_costumes_needed():
    mu = 161
    sigma = 4
    lower_bound = 152
    upper_bound = 158
    probability = norm.cdf(upper_bound, mu, sigma) - norm.cdf(lower_bound, mu, sigma)
    return probability

probability = probability_costumes_needed()
print("Доля костюмов для мальчиков с ростом от 152 до 158 см:", probability)
```

```python
# Задание 4
defective_products = 100
# Среднее количество бракованных деталей в день
defective_day = defective_products / 7

def probability_20_to_30_defective():
    lower_bound = 20
    upper_bound = 30
    probability = norm.cdf(upper_bound + 0.5, loc=defective_products, scale=(defective_products ** 0.5)) - norm.cdf(lower_bound - 0.5, loc=defective_products, scale=(defective_products ** 0.5))
    return probability

def probability_at_least_one_defective_per_day():
    probability = 1 - (1 - (1 / defective_day)) ** defective_day
    return probability

probability_20_to_30 = probability_20_to_30_defective()
probability_at_least_one = probability_at_least_one_defective_per_day()
print("Вероятность того, что за неделю будет от 20 до 30 бракованных деталей:", probability_20_to_30)
print("Вероятность того, что за день будет хотя бы одна забракованная деталь:", probability_at_least_one)
```

```python
# Задание 1

# Вероятность достать две качественные детали
p_qual_qual = (17 / (17 + 6)) * ((17 - 1) / (17 + 6 - 1))
# Вероятность достать одну качественную и одну бракованную деталь
p_qual_def = (17 / (17 + 6)) * (6 / (17 + 6 - 1)) + (6 / (17 + 6)) * (17 / (17 + 6 - 1))
# Вероятность достать две бракованные детали
p_def_def = (6 / (17 + 6)) * ((6 - 1) / (17 + 6 - 1))
# Прибыль для каждой комбинации
profit_qual_qual = 10 + 10
profit_qual_def = 10 - 20
profit_def_def = -20 - 20
# Средняя прибыль
expected_profit = p_qual_qual * profit_qual_qual + p_qual_def * profit_qual_def + p_def_def * profit_def_def
print("Средняя прибыль:", expected_profit, "рублей")
```
