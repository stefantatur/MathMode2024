# Графы

```python
# При помощи библиотеки NetworkX создайте ненаправленный граф, который состоит из 5 вершин и 4 ребер. Выведите такой граф на экран.
import networkx as nx
import matplotlib.pyplot as plt

graph = nx.Graph()
graph.add_nodes_from([1, 2, 3, 4, 5])
graph.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5)])
nx.draw(graph, with_labels=True)
plt.show()
```

```python
# При помощи библиотеки NetworkX создайте ненаправленный полный граф. Выведите такой граф на экран.
graph = nx.complete_graph(5)
nx.draw(graph, with_labels=True)
plt.show()
```

```python
# При помощи библиотеки NetworkX создайте ненаправленный связный граф. Выведите такой граф на экран.
graph = nx.connected_watts_strogatz_graph(5, 4, 0.5)
nx.draw(graph, with_labels=True)
plt.show()
```

```python
# При помощи библиотеки NetworkX создайте ненаправленный взвешенный полный граф, у которого 6 вершин (A,B,C,D,E,F). Веса рёбер такого графа пусть имеют значения 1,2,3,4,5 и т.д. по порядку. Выпишите кратчайшие пути между теми вершинами, которые имеют больше одного ребра на своём пути.
graph = nx.complete_graph(['A','B', 'C', 'D', 'E', 'F'])
weights = [i for i in range(1, len(graph.edges)+1)]
for (u, v), w in zip(graph.edges, weights):
    graph[u][v]['weight'] = w

shortest_paths = dict(nx.all_pairs_shortest_path(graph))
for source in shortest_paths:
    for target in shortest_paths[source]:
        if len(shortest_paths[source][target]) > 1:
            print(f"Shortest path from {source} to {target}: {shortest_paths[source][target]}")

nx.draw(graph, with_labels=True)
plt.show()
```
